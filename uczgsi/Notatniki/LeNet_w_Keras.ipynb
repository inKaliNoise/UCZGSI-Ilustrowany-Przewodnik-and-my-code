# LeNet w Keras

Ten notatnik implementuje głęboką sieć neuronową klasyfikującą cyfry MNIST. Inspiracją jest sieć [LeNet-5](http://yann.lecun.com/exdb/publis/pdf/lecun-01a.pdf).

#### Załadowanie zależności.


```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Flatten, Conv2D, MaxPooling2D # new!
```

#### Załadowanie danych.


```python
(X_train, y_train), (X_valid, y_valid) = mnist.load_data()
```

#### Wstępne przetworzenie danych.


```python
X_train = X_train.reshape(60000, 28, 28, 1).astype('float32')
X_valid = X_valid.reshape(10000, 28, 28, 1).astype('float32')
```

```python
X_train /= 255
X_valid /= 255
```

```python
n_classes = 10
y_train = keras.utils.to_categorical(y_train, n_classes)
y_valid = keras.utils.to_categorical(y_valid, n_classes)
```

#### Architektura sieci neuronowej.


```python
model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(n_classes, activation='softmax'))
```

```python
model.summary()
```

#### Konfiguracja modelu.


```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

#### Trening!


```python
model.fit(X_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(X_valid, y_valid))
```
