# Koszt kwadratowy

```python
def squared_error(y, yhat):
    return (y - yhat)**2
```

```python
squared_error(1, 1)
```

```python
squared_error(1, 0.9997)
```

```python
squared_error(1, 0.9)
```

```python
squared_error(1, 0.6)
```

```python
squared_error(1, 0.1192)
```

```python
squared_error(0, 0.1192)
```

```python
squared_error(1, 1-0.1192)
```
