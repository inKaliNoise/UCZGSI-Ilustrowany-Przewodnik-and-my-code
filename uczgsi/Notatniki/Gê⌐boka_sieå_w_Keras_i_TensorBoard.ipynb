# Głęboka sieć w Keras i TensorBoard

W tym notatniku wykorzystujemy kod głębokiej sieci, aby dostosować ją do narzędzia TensorBoard.

#### Załadowanie zależności.


```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout 
from keras.layers.normalization import BatchNormalization 
from keras.optimizers import SGD
from keras.callbacks import TensorBoard # Nowość! 
```

#### Załadowanie danych.


```python
(X_train, y_train), (X_valid, y_valid) = mnist.load_data()
```

#### Wstępne przetworzenie danych.


```python
X_train = X_train.reshape(60000, 784).astype('float32')
X_valid = X_valid.reshape(10000, 784).astype('float32')
```

```python
X_train /= 255
X_valid /= 255
```

```python
n_classes = 10
y_train = keras.utils.to_categorical(y_train, n_classes)
y_valid = keras.utils.to_categorical(y_valid, n_classes)
```

#### Architektura sieci neuronowej.


```python
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(BatchNormalization())

model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())

model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

#### Konfiguracja modelu.


```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

#### Katalog dla danych narzędzia TensorBoard.

```python
tensorboard = TensorBoard('logs/deep-net')
```

#### Trening!


```python
model.fit(X_train, y_train, 
          batch_size=128, 
          epochs=20, verbose=1, 
          validation_data=(X_valid, y_valid),
          callbacks=[tensorboard])
```
