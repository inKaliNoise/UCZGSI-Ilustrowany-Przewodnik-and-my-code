##### Kolokacje bigramów.

```python
phrases = Phrases(gberg_sents) # Trening detektora.
```

```python
bigram = Phraser(phrases) # Utworzenie wydajniejszego obiektu Phraser do przekształcania zdań.
```

```python
bigram.phrasegrams # Zliczenie wystąpień i ocena bigramów.
```

```python
bigram[tokenized_sentence]
```

#### Wstępne przetworzenie korpusu.

```python
lower_sents = []
for s in gberg_sents:
    lower_sents.append([w.lower() for w in s if w.lower()
                        not in list(string.punctuation)])
```

```python
lower_sents[0:5]
```

```python
lower_bigram = Phraser(Phrases(lower_sents))
```

```python
lower_bigram.phrasegrams
```

```python
lower_bigram["jon lives in new york city".split()]
```

```python
lower_bigram = Phraser(Phrases(lower_sents, 
                               min_count=32, threshold=64))
lower_bigram.phrasegrams
```

```python
clean_sents = []
for s in lower_sents:
    clean_sents.append(lower_bigram[s])
```

```python
clean_sents[0:9]
```

```python
clean_sents[6] 
```
