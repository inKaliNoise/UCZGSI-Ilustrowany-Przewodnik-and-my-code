```python
x = torch.zeros(3, 3, dtype=torch.float32, requires_grad=True)

y = x - 4
z = y**3 * 6
out = z.mean()

out.backward()

print(x.grad)
```

```python
# Zdefiniowanie losowych tensorów dla wartości wejściowych i wyjściowych.
x = torch.randn(32, 784, requires_grad=True)
y = torch.randint(low=0, high=10, size=(32,))



# Zdefiniowanie modelu za pomocą klasy Sequential.
model = torch.nn.Sequential(
    torch.nn.Linear(784, 100),
    torch.nn.Sigmoid(),
    torch.nn.Linear(100, 10),
    torch.nn.LogSoftmax(dim=1)
)

# Zdefiniowanie optymalizatora i funkcji straty.
optimizer = torch.optim.Adam(model.parameters())
loss_fn = torch.nn.NLLLoss()

for step in range(1000):
    # Prognozowanie z wykorzystaniem propagacji w przód.
    y_hat = model(x)
    # Wyliczenie straty.
    loss = loss_fn(y_hat, y)
    # Wyzerowanie gradientu przed wykonaniem przebiegu wstecz.
    optimizer.zero_grad()
    # Wyliczenie gradientów względem straty.
    loss.backward()
    # Wyświetlenie wyników.
    print('Krok: :4d, strata: :0.4f'.format(step+1, loss.item()))
    # Aktualizacja parametrów modelu.
    optimizer.step()
```
